```python
#########################
### Train & Orig Table ##

def table(df1, df2, feat, n1='tr', n2='te'):
    num_feat = df1[feat].select_dtypes(include='number').columns.tolist()
    if num_feat:
        df1_d = df1[num_feat].describe().transpose()
        df2_d = df2[num_feat].describe().transpose()
    table = pd.DataFrame(index=feat)
    table['Type']     = df1[feat].dtypes
    table[f'{n1}Nul'] = df1[feat].isnull().sum()
    table[f'{n2}Nul'] = df2[feat].isnull().sum()
    table[f'{n1}Uni'] = df1[feat].nunique()
    table[f'{n2}Uni'] = df2[feat].nunique()
    if num_feat:
        table[f'{n1}Avg'] = df1_d['mean'].round(1)
        table[f'{n2}Avg'] = df2_d['mean'].round(1)
        table[f'{n1}Min'] = df1_d['min'].round(1)
        table[f'{n2}Min'] = df2_d['min'].round(1)
        table[f'{n1}Max'] = df1_d['max'].round(1)
        table[f'{n2}Max'] = df2_d['max'].round(1)
    int_metric = [f'{n1}Nul', f'{n2}Nul', f'{n1}Uni', f'{n2}Uni']
    table[int_metric] = table[int_metric].fillna(0).astype(int)
    table = table.sort_values(by=f'{n1}Uni')
    return table

all_features = train.drop(columns=[target]).columns.tolist()
table(train, orig, all_features, 'tr', 'or')
```

```python
#########################
### Train & Test Table ##

train['extra'] = 0
orig['extra']  = 1
test['extra']  = 0
print('raw train:',train.shape)
print('raw orig :',orig.shape)

raw   = len(train) + len(orig)
train = pd.concat([train, orig],axis=0)
check = train.drop(columns=['extra']).columns.tolist()
train.drop_duplicates(subset=check, inplace=True)
new = len(train)

print('----- ----- ----- -----')
print('train_orig duplicate:', raw-new)
print('----- ----- ----- -----')
print('new train:',train.shape)
print('new test :',test.shape)
print('----- ----- ----- -----')

table(train, test, all_features, 'tr', 'te')
```

```python
#########################
### Numeric Cleaning ####

def num_preprocess(df):
    df['Duration']   = df['Duration'].astype(int)
    df['Age']        = df['Age'].astype(int)
    df['Heart_Rate'] = np.round(df['Heart_Rate'],0).astype(int)
    df['Body_Temp']  = np.round(df['Body_Temp']*10,0).astype(int)
    df['Height']     = df['Height'].clip(127,219).astype(int)
    df['Weight']     = df['Weight'].clip(39,126).astype(int)
    return df
df2 = num_preprocess(df2)
num_features = df2[num_features].nunique().sort_values().index.tolist()

table(df2[:len(train)], df2[len(train):], num_features)
```

```python
#########################
### Feature Interaction #

fi_base = df2.columns[df2.nunique() >= 10].drop([target]).tolist()
fi_base = df2[fi_base].nunique().sort_values().index.tolist()
fi_feat = []

combo = list(combinations(fi_base, 2))
for col1, col2 in tqdm(combo):
    feat_mul = f'{col1}_{col2}_mul'
    feat_div = f'{col1}_{col2}_div'
    df3[feat_mul] = df3[col1] * df3[col2]
    df3[feat_div] = df3[col1] / df3[col2]
    fi_feat.extend([feat_mul, feat_div])
gc.collect()
fi_feat
```

```python
#########################
### K-Folds Spliting ####

def Split(df3, train, n=5, seed=seed, print_info=True):
    Train = df3.iloc[:len(train)]
    Test  = df3.iloc[len(train):].drop(columns=[target])
    tra_x = Train[Train['extra'] == 0].drop(columns=[target])
    tra_y = Train[Train['extra'] == 0][target]
    ext_x = Train[Train['extra'] == 1].drop(columns=[target])
    ext_y = Train[Train['extra'] == 1][target]
    folds, oof_true = [], []    
    kfold = StratifiedKFold(n_splits=n, shuffle=True, random_state=seed)
    for tr_idx, va_idx in kfold.split(tra_x, tra_y.astype(str)):
        x_train = tra_x.iloc[tr_idx]
        y_train = tra_y.iloc[tr_idx]
        x_valid = tra_x.iloc[va_idx]
        y_valid = tra_y.iloc[va_idx]
        x_train = pd.concat([x_train, ext_x], axis=0).reset_index(drop=True)
        y_train = pd.concat([y_train, ext_y], axis=0).reset_index(drop=True)
        x_test  = Test.copy()
        '''
        T = TargetEncoder(n_folds=5, seed=seed, stat='mean')
        for col in tqdm(te_feat):
            x_train[col]=T.fit_transform(x_train[[col]],y_train).astype('float32')
            x_valid[col]=T.transform(x_valid[[col]]).astype('float32')
            x_test[col] =T.transform(x_test[[col]]).astype('float32')
        '''
        folds.append((x_train, x_valid, y_train, y_valid, x_test))
        oof_true.extend(y_valid)
    oof_table = pd.DataFrame({'oof_target': oof_true})
    pred_table = pd.DataFrame()
    if print_info:
        print('x_train', x_train.shape)
        print('x_valid', x_valid.shape)
        print('y_train', y_train.shape)
        print('y_valid', y_valid.shape)
        print('x_test ', x_test.shape)        
    return folds, oof_true, oof_table, pred_table

folds, oof_true, oof_table, pred_table = Split(df3, train, n=5)
gc.collect()
```

```python
#########################
### XGBoost Impotance ###

def Model(best_param, folds=folds, oof_true=oof_true, verbose=True):
    oof, pred, n = [], 0, len(folds)
    boosters = []
    features = folds[0][0].columns.tolist()
    importance_dict = {f: 0.0 for f in features}
    for i, (x_train, x_valid, y_train, y_valid, x_test) in enumerate(folds):
        xgb_train = xgb.DMatrix(x_train, y_train)
        xgb_valid = xgb.DMatrix(x_valid, y_valid)
        model = xgb.train(best_param, xgb_train, 20000,evals=[(xgb_valid,"")],
                          early_stopping_rounds=50, verbose_eval=0)
        oof_f = model.predict(xgb_valid)
        oof.extend(oof_f)
        pred += model.predict(xgb.DMatrix(x_test))/n
        boosters.append(model)
        score = model.get_score(importance_type='gain')
        for feat in features:
            importance_dict[feat] += score.get(feat, 0)
        metric_fold = mean_squared_error(y_valid, oof_f, squared=False)
        if verbose:
            print(f'Fold {i+1}: Metric: {metric_fold:.5f}')
    metric_overall = mean_squared_error(oof_true, oof, squared=False)
    print(f'Overall Metric: {metric_overall:.5f}')
    sorted_feat = sorted(importance_dict.items(), key=lambda x: -x[1])
    feat_importance = [f[0] for f in sorted_feat]
    return oof, pred, feat_importance

param = {'learning_rate'   : 0.03, 
         'reg_alpha'       : 0.3,
         'reg_lambda'      : 0.3,
         'subsample'       : 0.7,
         'colsample_bytree': 0.7,
         'random_state'    : 0,
         'device'          :'cuda', 
         'objective'       :'reg:squarederror', 
         'eval_metric'     :'rmse'}
start = time.time()
oof, pred, importance_list = Model(param, verbose=True)
end = time.time()
print(f'Time: {end - start:.2f} seconds')
gc.collect()
```

```python
#########################
### LGB & XGB & CAT #####

def Model(model_name, best_param, folds=folds, oof_true=oof_true, verbose=True):
    oof, pred, n = [], 0, len(folds)
    for i, (x_train, x_valid, y_train, y_valid, x_test) in enumerate(folds):
        if model_name == 'lgb':
            lgb_train = lgb.Dataset(x_train,y_train,categorical_feature=cat_feat)
            lgb_valid = lgb.Dataset(x_valid,y_valid,categorical_feature=cat_feat)
            model = lgb.train(best_param,lgb_train,20000,valid_sets=[lgb_valid], 
                              callbacks=[lgb.early_stopping(500, verbose=0)])
            oof_f = model.predict(x_valid)
            pred += model.predict(x_test)/n
        elif model_name == 'xgb':
            xgb_train = xgb.DMatrix(x_train, y_train)
            xgb_valid = xgb.DMatrix(x_valid, y_valid)
            model = xgb.train(best_param, xgb_train, 20000,evals=[(xgb_valid,"")],
                              early_stopping_rounds=500, verbose_eval=0)
            oof_f = model.predict(xgb_valid)
            pred += model.predict(xgb.DMatrix(x_test))/n
        elif model_name == 'cat':
            cat_train = cat.Pool(x_train, y_train, cat_features=cat_feat)
            cat_valid = cat.Pool(x_valid, y_valid, cat_features=cat_feat)
            model = cat.train(cat_train, best_param, eval_set=[cat_valid], 
                    iterations=20000, early_stopping_rounds=500, verbose=0)
            oof_f = model.predict(x_valid)
            pred += model.predict(x_test)/n
        metric_fold = mean_squared_error(y_valid, oof_f, squared=False)
        if verbose:
            print(f'Fold {i+1}: Metric: {metric_fold:.5f}')
        oof.extend(oof_f)
    metric_overall = mean_squared_error(oof_true, oof, squared=False)
    print(f'Overall Metric: {metric_overall:.5f}')
    return oof, pred
```

```python
#########################
### LGB & XGB & CAT #####

def Model(model_name, best_param, folds=folds, oof_true=oof_true, verbose=True):
    oof, pred, n = [], 0, len(folds)
    for i, (x_train, x_valid, y_train, y_valid, x_test) in enumerate(folds):
        if model_name == 'lgb':
            lgb_train = lgb.Dataset(x_train,y_train,categorical_feature=cat_feat)
            lgb_valid = lgb.Dataset(x_valid,y_valid,categorical_feature=cat_feat)
            model = lgb.train(best_param,lgb_train,20000,valid_sets=[lgb_valid], 
                              callbacks=[lgb.early_stopping(500, verbose=0)])
            oof_f = model.predict(x_valid)
            pred += model.predict(x_test)/n
        elif model_name == 'xgb':
            xgb_train = xgb.DMatrix(x_train, y_train)
            xgb_valid = xgb.DMatrix(x_valid, y_valid)
            model = xgb.train(best_param, xgb_train, 20000,evals=[(xgb_valid,"")],
                              early_stopping_rounds=500, verbose_eval=0)
            oof_f = model.predict(xgb_valid)
            pred += model.predict(xgb.DMatrix(x_test))/n
        elif model_name == 'cat':
            cat_train = cat.Pool(x_train, y_train, cat_features=cat_feat)
            cat_valid = cat.Pool(x_valid, y_valid, cat_features=cat_feat)
            model = cat.train(cat_train, best_param, eval_set=[cat_valid], 
                    iterations=20000, early_stopping_rounds=500, verbose=1000)
            oof_f = model.predict(x_valid)
            pred += model.predict(x_test)/n
        metric_fold = mean_squared_error(y_valid, oof_f, squared=False)
        if verbose:
            print(f'Fold {i+1}: Metric: {metric_fold:.5f}')
        oof.extend(oof_f)
    metric_overall = mean_squared_error(oof_true, oof, squared=False)
    print(f'Overall Metric: {metric_overall:.5f}')
    return oof, pred
```
